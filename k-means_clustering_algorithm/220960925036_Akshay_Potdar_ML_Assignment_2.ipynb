Implement k-means algorithm using python program:
Sample dataset: 2,4,10,12,3,20,30,11,25
Take first mean as 2 and 4

```python
#declaring an set of elements
data={2,4,10,12,3,20,30,11,25}
df=list(data)
#Mean Assignment
m1=2
m2=4
#Create two list to store clusters
l1=[]
l2=[]
#Comparing the lists
temp1,temp2=[0,],[1,]
```

```python
#Creating Two clusters
def cluster_maker(dataframe,m1=0,m2=0):
    k1,k2=list(), list()
    d1,d2=0,0
    if m1==0 and m2==0:
        m1,m2=dataframe[0],dataframe[1,]
    #Finding distance
    for i in dataframe:
        d1 = abs(i-m1)
        d2 = abs(i-m2)
        if(d1<=d2):
            k1.append(i)
    else:
            k2.append(i)
    return k1,k2
```

```python
def listavg(dummylist):
    sum, avg = 0,0.0
    for i in dummylist:
        sum+=i
    avg = sum/len(dummylist)
    return avg

while(l1!=temp1):
    temp1,temp2 = cluster_maker(df,m1,m2)
    m1 = listavg(temp1)
    m2 = listavg(temp2)
    l1,l2 = cluster_maker(df,m1,m2)
    
print("Cluster 1 : ",l1,"\nCluster 2 :",l2)
```
